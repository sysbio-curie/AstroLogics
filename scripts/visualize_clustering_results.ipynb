```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
rand_index= pd.read_csv('/home/spankaew/Git/astrologics/tmp/rand_index_endpointclustering.csv')
rand_endpoint = np.trapezoid(rand_index, axis = 0)
plt.figure(figsize=(20, 10))
sns.boxplot(data=rand_index.transpose(), orient='v')
plt.xlabel('Values')
plt.ylabel('Index')
plt.title('Box Plot of rand_index')
plt.show()
```

```python
rand_index= pd.read_csv('/home/spankaew/Git/astrologics/tmp/rand_index_alltrajclustering.csv')
rand_alltraj = np.trapezoid(rand_index, axis = 0)
plt.figure(figsize=(20, 10))
sns.boxplot(data=rand_index.transpose(), orient='v')
plt.xlabel('Values')
plt.ylabel('Index')
plt.title('Box Plot of rand_index')
plt.show()
```

```python
# Convert numpy arrays to pandas DataFrame
rand_endpoint_df = pd.DataFrame(rand_endpoint, columns=['rand_endpoint'])/74
rand_alltraj_df = pd.DataFrame(rand_alltraj, columns=['rand_alltraj'])/74

# Concatenate the DataFrames
result = pd.concat([rand_endpoint_df, rand_alltraj_df], axis=1)
result.columns = ['Endpoint', 'Whole Trajectories']
# Create a boxplot comparing the distribution of the two columns
plt.figure(figsize=(6, 6))
sns.boxplot(data=result, orient='v', width = .5)
#plt.xlabel('Clustering Method')
plt.ylabel('Rand Index')
plt.title('Comparison of average rand index: \n Endpoint vs Whole Trajectories')
plt.ylim(0.6,0.8)
plt.grid()
plt.savefig('/home/spankaew/Git/astrologics/figures/BNtuto_rand_index_comparison.png', bbox_inches='tight', dpi=600)
plt.savefig('/home/spankaew/Git/astrologics/figures/BNtuto_rand_index_comparison.pdf', bbox_inches='tight')
plt.show()
```
